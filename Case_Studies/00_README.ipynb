# Inductive Biases of ML Architectures — a Seven-Notebook Tour

**One idea runs through every notebook:** an architecture is an *assumption about what structure the data has* (its **inductive bias**). A model does extremely well **exactly when the data-generating process matches its bias**, and is blind in the complementary direction. Each notebook (1) states the worldview, (2) **synthesizes data that matches it** to show an "extreme good" case against ground truth, and (3) shows a **blind-spot** case where the assumption is wrong — and that blind spot is precisely what the *next* architecture is built to fix.

Every code cell is self-contained, verbose-commented, and prints metrics plus figures you can reason about.

| # | Notebook | Worldview | Extreme good case (result) | Blind spot (result) |
|---|----------|-----------|----------------------------|---------------------|
| 01 | Dynamic programming | biology is a **string**; similarity = edit cost | homolog vs random separation, **AUC 1.00** | reverse-complement relation invisible, **AUC ≈ 0.60** |
| 02 | Hidden Markov models | a **stochastic process with hidden states** | hidden-state recovery **≈ 91%** + emission mechanism recovered | lag-20 dependency sits at the i.i.d. floor (non-Markovian) |
| 03 | SVMs & kernels | data **embeds in a feature space**; kernel = similarity | RBF on rings **acc 1.00** vs linear 0.53; k-mer string kernel **AUC 0.81** | wrong RBF γ swings accuracy 0.54 → 1.00 (kernel *is* the assumption) |
| 04 | Tree ensembles | biology is **tabular**; signal in feature interactions | GB **AUC 0.74** vs logistic 0.58; 3 real features recovered among 43 | translation-invariant signal defeats axis-aligned splits |
| 05 | CNNs | **translation-invariant local motifs** | 1-filter CNN **AUC 0.98** > GB 0.83; transfers to unseen positions **0.99**; filter is a motif detector **AUC 0.97** | absolute-position signal → **AUC 0.47** (pooled position away) |
| 06 | Transformers / attention | **long-range dependencies**, learned | distant-pair task **AUC 1.00** vs local CNN 0.52; attention map lights up the true pair; distance-flat across gaps | O(n²) compute + data-hungry; CNN wins a small-data **local** task |
| 07 | Graph neural networks | biology is **relational** (a network) | community labels from edges alone, GNN **0.91** vs MLP 0.56 | corrupted graph → chance; **over-smoothing** collapses depth ≥ 8 |

## How to run
Each notebook is independent. Dependencies: `numpy`, `scipy`, `scikit-learn`, `matplotlib`, `seaborn`, `hmmlearn` (notebook 02), and `torch` (notebooks 05–07, CPU is fine). Run top to bottom; every figure regenerates from synthetic data with fixed seeds.

## The lineage, in one line
DP assumes strings → HMMs add a learned stochastic latent process → SVMs assume a similarity geometry → trees assume tabular interactions → CNNs assume translation-invariant motifs → transformers assume learnable long-range structure → GNNs assume relational structure. **Each blind spot motivates the next architecture — which is why the strongest modern systems *combine* these biases rather than picking one.**
